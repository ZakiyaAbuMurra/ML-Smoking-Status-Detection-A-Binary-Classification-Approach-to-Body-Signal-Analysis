```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import  GridSearchCV
from google.colab import files
import xgboost as xgb
from sklearn.model_selection import train_test_split
```

```python
train_set = pd.read_csv('/content/trainClean.csv')
test_set = pd.read_csv('/content/testClean.csv')
val_set = pd.read_csv('/content/valClean.csv')
```

```python
dataSet = pd.read_csv('/content/cleanDataSet.csv')
dataSet
```

```python
stat = dataSet.describe().round(2).transpose()
stat = stat
```

```python
X_train = train_set.iloc[:, :-1]
y_train = train_set.iloc[:, -1]
X_test = test_set.iloc[:, :-1]
y_test = test_set.iloc[:, -1]
X_val = val_set.iloc[:, :-1]
y_val = val_set.iloc[:, -1]
```

**KNN with k =1**

```python


# Train the k-NN model
knn = KNeighborsClassifier(n_neighbors= 1)  # starting with k=5
knn.fit(X_train, y_train)

# Predict on the validation set
y_val_pred = knn.predict(X_val)

y_test_pred = knn.predict(X_test)


# # Evaluate the model
print("Validation Set Evaluation:")
print("Confusion Matrix:\n", confusion_matrix(y_val, y_val_pred))
print("Classification Report:\n", classification_report(y_val, y_val_pred))
print("Accuracy:", accuracy_score(y_val, y_val_pred))


print("Test Set Evaluation:")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred))
print("Classification Report:\n", classification_report(y_test, y_test_pred))
print("Accuracy:", accuracy_score(y_test, y_test_pred))
```

**KNN with K=3**

```python


# Train the k-NN model
knn = KNeighborsClassifier(n_neighbors= 3)  # starting with k=5
knn.fit(X_train, y_train)

# Predict on the validation set
y_val_pred = knn.predict(X_val)

y_test_pred = knn.predict(X_test)


# # Evaluate the model
print("Validation Set Evaluation:")
print("Confusion Matrix:\n", confusion_matrix(y_val, y_val_pred))
print("Classification Report:\n", classification_report(y_val, y_val_pred))
print("Accuracy:", accuracy_score(y_val, y_val_pred))


print("Test Set Evaluation:")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred))
print("Classification Report:\n", classification_report(y_test, y_test_pred))
print("Accuracy:", accuracy_score(y_test, y_test_pred))
```

```python
num_negative_class = sum(y_train == 0)
num_positive_class = sum(y_train == 1)

class_ratio = num_negative_class / num_positive_class

print(f'Class Ratio: {class_ratio}')
```
